```python
# Load Data csv file
import pandas as pd
dataset=pd.read_csv('Position_Salaries.csv')
```

```python
# Descriptive Head of dataset
print("Descriptive Head of dataset")
print(dataset.head(4))
# Descriptive shape of dataset
print("Descriptive shape of dataset")
print(dataset.shape)
# Descriptive datatype of each attribute
print("Descriptive datatype of each attribute")
print(dataset.dtypes)
```

```python
# Visualizating the each attribute in histogram and density plots
import matplotlib.pyplot as plt
# Histogram
dataset.hist(layout=(2,2))
plt.show()
```

```python
# Density plot to view smooth curve for each attribute
dataset.plot(kind='density',sharex=False,sharey=False,subplots=True,layout=(2,2))
plt.show()
```

```python
# As per above Visualization of each attribute both input and output attribute are having
# Gaussian Distribution
```

```python
# Split The dataset into input and output attribute
array=dataset.values
x=array[:,1:2]
y=array[:,-1]
print("------------------ x input attribute----------")
print(x[:4,:])
print("-------------------y output attribute---------")
print(y[:4,])
```

```python
# We are not splitting dataset into train and test set because we have very less no of data avilable
```

```python
# Simple linear regression 
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x,y)
```

```python
# Polynomial Regression Algorithm with Linear regression
# In Polynomial regression algorithm we just bending the predicted curve with some degree
# depends on degree value we will get to know normal to overfitting
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
poly.fit(x_poly,y)
regression2=LinearRegression()
regression2.fit(x_poly,y)
```

```python
# y_pred for trained dataset x_poly
y_pred=regression2.predict(x_poly)
print(y_pred)
```

```python
# Visualising linear regression result
plt.scatter(x,y,color='blue')
plt.plot(x,regression.predict(x),color='red')
plt.title('Linear Regression Result')
plt.xlabel('Year Of Experience')
plt.ylabel('Salarys(K$)')
plt.show()
```

```python
# As Per linear Regression Algorithm Doesn't fit the data properly it comes "UnderFit"
```

```python
# Visualising Polynomial Regression Algorithm  Result
plt.scatter(x,y,color='blue')
plt.plot(x,regression2.predict(poly.fit_transform(x)),color='red')
plt.title('Polynomial Regression Algorithm')
plt.xlabel("Year Of Experience")
plt.ylabel('Salary(K$)')
plt.show()
```

```python
# As per Above Visualisation Polynomial regression algorithm "overfitting".
# It looks like pretty overfit because we have less data and we didnt predict the unseen data.
# We Predicted with trained data
```
