```python
# Hierarchical Cluster Algorithm For Mall Customer Dataset
```

```python
# Load csf file

import pandas as pd
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
# Descriptive head of a dataset
print(dataset.head(4))
# Descriptive shape of a dataset
print(dataset.shape)
# Descriptive datatype for each attribute in a dataset
print(dataset.dtypes)
```

```python
# Visualistion plot for each attribute to know what kind of distribution it is?

import matplotlib.pyplot as plt
# Histogram plot
dataset.hist()
plt.show()
```

```python
# To view plots smoothly we are using density plot

# Density plot
dataset.plot(kind='density',sharex=False,sharey=False,subplots=True,layout=(2,2))
plt.show()
```

```python
# Split the dataset into input attributes and not accesing the irrelvant feature attributes
x=dataset.iloc[:,3:5].values
print(x[:5,:])
```

```python
# Using Dendogram finding optimal number of cluster for Hierarchical cluster
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('customers')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
# As per the above dendrogram figure. long region didnt cross any line.
# So optimal no.of clusters are "5".
```

```python
# Hierarchical Clustering Algorithm to the mall dataset
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5)
hc.fit(x)
```

```python
# Predict the cluster categories based on mall dataset
y_hc=hc.fit_predict(x)
print(y_hc)
```

```python
# As we can see each row customer data belong to one cluster based on those cluster we will predict 
# how much that customer valuble to mall future purchase.
# based on their cluster categories mall give some extra befits to their customer to keep them as a valuable customers.
```
