```python
# Decision Tree Algorithm To predict furtur employee salary based on his/her position
```

```python
# Load csv file
import pandas as pd
dataset=pd.read_csv('Position_Salaries.csv')
```

```python
# Descriptive head of a dataset
print("Descriptive head of a dataset",dataset.head(4))
# Descriptive shape of a dataset
print("Descriptive shape of a dataset",dataset.shape)
# Descriptive data type of a dataset
print("Descriptive data type of a dataset",dataset.dtypes)
```

```python
# Visualisation of each attribute

import matplotlib.pyplot as plt
# Histogram plot of an each attibute
dataset.hist()
plt.show()
```

```python
# Density Visualisation for each attribute
dataset.plot(kind='density',subplots=True,sharex=False,sharey=False,layout=(2,2))
plt.show()
```

```python
# split the dataset into input and output attribute
array=dataset.values
x=array[:,1:2]
y=array[:,-1]
```

```python
# Decision tree Regression Algorithm
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x,y)
```

```python
# Predicting model
y_pred=dtr.predict(x)
print(y_pred)
```

```python
# Accuracy Metrics of an algorithm
from math import sqrt
def root_mean_square_error(actual,predicted):
    sum_error=0.0
    for i in range(len(actual)):
        yhat=predicted[i]-actual[i]
        sum_error+=(yhat**2)
    rmse=sum_error/float(len(actual))
    return sqrt(rmse)
```

```python
# Accuracy of model
accuracy=root_mean_square_error(x,dtr.predict(x))
print("accuracy of an model",accuracy)
```

```python
# Visualising result predicted by model
plt.scatter(x,y,color='blue')
plt.plot(x,dtr.predict(x),color='red')
plt.title(" Decision Tree Regression Result")
plt.xlabel("Year Of Experience")
plt.ylabel("salary")
plt.show()
```
