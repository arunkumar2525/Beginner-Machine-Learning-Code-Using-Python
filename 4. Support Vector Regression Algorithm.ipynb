```python
# Support Vector Regression Algorithm For Future Position Employee Salary Predict Dataset
```

```python
# Load CSV file
import pandas as pd
dataset=pd.read_csv('Position_Salaries.csv')
```

```python
# Descriptive head of dataset
print("Descriptive Head of dataset",dataset.head(4))
# Descriptive Shape of dataset
print("Descriptive Shape of dataset",dataset.shape)
# Descriptive Datatype of dataset
print("Descriptive Datatype of dataset",dataset.dtypes)
```

```python
# Visualisation of each attribute to understand what kind of distribution it is

import matplotlib.pyplot as plt
# Histogram plot for each attribute
dataset.hist()
plt.show()
```

```python
# Density Visualization of each attribute
dataset.plot(kind='density',subplots=True,sharex=False,sharey=False,layout=(2,2))
plt.show()
```

```python
# split the dataset into input and output attribute
array=dataset.values
x=array[:,1:2]
y=array[:,-1]
print("------------------ x Input Attribute-----------------")
print(x[:4,:])
print("------------------- y Output Attribute --------------")
print(y[:4,])
```

```python
# Not Splitting the dataset into training and test set
```

```python
# Support Vector Regression Algorithm
from sklearn.svm import SVR
support=SVR(kernel='poly',degree=5)
support.fit(x,y)
```

```python
# Predicting the accuracy of a model we are using seen data to predict accuracy
y_pred=support.predict(x)
print(y_pred)
```

```python
# Accuracy metrics of an algorithm
from math import sqrt
def root_mean_squared_error(actual,predicted):
    sum_error=0.0
    for i in range(len(actual)):
        yhat=predicted[i]-actual[i]
        sum_error+=(yhat**2)
    rmse=sum_error/float(len(actual))
    return sqrt(rmse)
```

```python
# Accuracy metrics of an algorithm
accuracy=root_mean_squared_error(x,support.predict(x))
print("Accuracy of an model",accuracy)
```

```python
# Visualising the Result in dataset
plt.scatter(x,y,color='blue')
plt.plot(x,support.predict(x),color='red')
plt.title("Support Vector Regression")
plt.xlabel("Year Of Experience")
plt.ylabel("Salary")
plt.show()
```
